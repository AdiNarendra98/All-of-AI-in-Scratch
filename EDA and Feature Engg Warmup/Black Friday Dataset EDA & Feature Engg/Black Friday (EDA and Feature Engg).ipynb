```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# IMPORTING TRAIN DATASET
df_train= pd.read_csv('train.csv')
df_train.head()
```

```python
# IMPORTING TEST DATASET
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
# MERGE or APPEND BOTH TEST AND TRAIN DATA:
df=df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Deleting Unecessary Features
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# CONVERISION OF CATERGORICAL feature(GENDER) TO NUMERICAL:
df["Gender"]=pd.get_dummies(df["Gender"],drop_first=1)
#This is Time consuming to add pd to df.So We use amnother method
df.head()
```

```python
df['Age'].unique()
```

```python
# CONVERISION OF CATERGORICAL feature(AGE) TO NUMERICAL:1st method
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head()
```

```python
# CONVERISION OF CATERGORICAL feature(City_Category) TO NUMERICAL:
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
df.drop('City_Category',axis=1,inplace=True) #Dropping City Feature after conversion
df.head()
```

```python
#MISSING VALUES FINDING:
df.isnull().sum()
```

```python
#Focus on Replacing Missing Values:
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
df['Product_Category_2'].mode()[0]
```

```python
#Replacing Missing Values with Mode:
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
df['Product_Category_3'].unique()
```

```python
#Replacing Missing Values with Mode:
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
df.info()
```

```python
#Convert Object feature to Integers:
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
df.info()
```

```python
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
df['Gender']=df['Gender'].astype(int)
df.info()
```
